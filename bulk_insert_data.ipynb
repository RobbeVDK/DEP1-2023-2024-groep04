```python
import pandas as pd;

```

```python
doelpunten_df = pd.read_csv('./csv/doelpunten_onverwerkd.csv')
klassement_df = pd.read_csv('./csv/klassement_onverwerkd.csv')
wedstrijden_df = pd.read_csv('./csv/wedstrijden_onverwerkd.csv')
```

```python
df_total = pd.merge(wedstrijden_df,doelpunten_df,on=['id_match','datum','seizoen','tijd','thuisploeg','uitploeg','uitploeg_stamnummer','thuisploeg_stamnummer'],how='left')
```

```python
rows_home = df_total[['thuisploeg', 'thuisploeg_stamnummer']].rename(columns={'thuisploeg': 'TeamName', 'thuisploeg_stamnummer': 'StamNumber'})
rows_away = df_total[['uitploeg', 'uitploeg_stamnummer']].rename(columns={'uitploeg': 'TeamName', 'uitploeg_stamnummer': 'StamNumber'})

dim_team_df = pd.concat([rows_home,rows_away])
dim_team_df = dim_team_df.drop_duplicates()
dim_team_df = dim_team_df.reset_index(drop=True)
dim_team_df['teamId'] = dim_team_df.index
dim_team_df
```

```python
import pandas as pd

dim_standings_df = klassement_df[['speeldag', 'seizoen', 'id_ploeg', 'stand', 'punten', 'aantal_gewonnen', 'aantal_gelijk', 'aantal_verloren', 'verschil']]
temp_df = dim_team_df.copy()
temp_df['ClubID'] = temp_df.index
merged_df = dim_standings_df.merge(temp_df, left_on='id_ploeg', right_on='StamNumber', how='left')

merged_df = merged_df.drop(columns=['id_ploeg','TeamName','StamNumber'])
dim_standings_df = merged_df.rename(columns={'speeldag':'StandingsPlayday','seizoen':'StandingsYear','stand':'Ranking','punten':'Points','aantal_gewonnen':'Wins','aantal_gelijk':'Ties','aantal_verloren':'Losses','verschil':'GoalDifference'})
dim_standings_df['StandingsDayId'] = dim_standings_df.index
dim_standings_df
```

```python
import pandas as pd


dim_date_df = pd.DataFrame()
dim_date_df['FullDate'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d')
dim_date_df['Year'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.year
dim_date_df['Quarter'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.quarter
dim_date_df['Month'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.month
dim_date_df['Day'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.day

dim_date_df = dim_date_df.drop_duplicates()
dim_date_df = dim_date_df.reset_index(drop=True)
dim_date_df['DateKey'] = dim_date_df.index
dim_team_df
dim_date_df
```
