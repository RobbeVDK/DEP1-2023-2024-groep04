```python
import pandas as pd;

```

```python
doelpunten_df = pd.read_csv('./csv/doelpunten.csv',sep=';',names=['seizoen', 'dag', 'datum', 'tijd', 'id_match', 'thuisploeg_stamnummer',
       'thuisploeg', 'uitploeg', 'uitploeg_stamnummer', 'goal_time',
       'real_time_goal', 'goal_team_naam', 'goal_team_stamnummer',
       'stand_thuis', 'stand_uit'])
klassement_df = pd.read_csv('./csv/klassement.csv',sep=';',names=['seizoen', 'speeldag', 'stand', 'id_ploeg', 'naam_ploeg',
       'aantal_wedstrijden', 'aantal_gewonnen', 'aantal_gelijk',
       'aantal_verloren', 'doelpunten_voor', 'doelpunten_tegen', 'verschil',
       'punten_2p_links', 'punten_2p_rechts', 'punten_3p'])
wedstrijden_df = pd.read_csv('./csv/wedstrijden.csv',sep=';',names=['seizoen', 'speeldag', 'datum', 'tijd', 'id_match',
       'thuisploeg_stamnummer', 'thuisploeg', 'uitploeg_stamnummer',
       'uitploeg', 'doelpunten_thuisploeg', 'doelpunten_uitploeg'])
```

```python
wedstrijden_df.columns
df_total = pd.merge(wedstrijden_df,doelpunten_df,on=['id_match','datum','seizoen','tijd','thuisploeg','uitploeg','uitploeg_stamnummer','thuisploeg_stamnummer'])
df_total
```

```python
rows_home = df_total[['thuisploeg', 'thuisploeg_stamnummer']].rename(columns={'thuisploeg': 'TeamName', 'thuisploeg_stamnummer': 'StamNumber'})
rows_away = df_total[['uitploeg', 'uitploeg_stamnummer']].rename(columns={'uitploeg': 'TeamName', 'uitploeg_stamnummer': 'StamNumber'})

dim_team_df = pd.concat([rows_home,rows_away])
dim_team_df = dim_team_df.drop_duplicates()
dim_team_df = dim_team_df.reset_index(drop=True)
dim_team_df['TeamKey'] = dim_team_df.index
dim_team_df
```

```python
import pandas as pd


dim_date_df = pd.DataFrame()
dim_date_df['FullDate'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d')
dim_date_df['Year'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.year
dim_date_df['Quarter'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.quarter
dim_date_df['Month'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.month
dim_date_df['Day'] = pd.to_datetime(wedstrijden_df['datum'], format='%Y/%m/%d').dt.day

dim_date_df = dim_date_df.drop_duplicates()
dim_date_df = dim_date_df.reset_index(drop=True)
dim_date_df['DateKey'] = dim_date_df.index
dim_date_df
```

```python
dim_time_df = pd.DataFrame()
dim_time_df['Hour'] = pd.to_datetime(wedstrijden_df['tijd'], format="%H:%M").dt.hour
dim_time_df['Minute'] = pd.to_datetime(wedstrijden_df['tijd'], format="%H:%M").dt.minute

dim_time_df = dim_time_df.drop_duplicates()
dim_time_df = dim_time_df.reset_index(drop=True)
dim_time_df['TimeKey'] = dim_time_df.index
dim_time_df


```

```python
import pandas as pd

dim_standings_df = klassement_df[['speeldag', 'seizoen', 'id_ploeg', 'stand', 'punten_3p', 'aantal_gewonnen', 'aantal_gelijk', 'aantal_verloren', 'verschil']]

temp_df = wedstrijden_df.merge(klassement_df,on=['seizoen','speeldag'])
temp_df['datum'] = pd.to_datetime(temp_df['datum'])
temp_df = temp_df.groupby(['seizoen','speeldag'])['datum'].max()
dim_standings_df = dim_standings_df.rename(columns={'speeldag':'StandingsPlayday','seizoen':'StandingsYear','stand':'Ranking','punten_3p':'Points','aantal_gewonnen':'Wins','aantal_gelijk':'Ties','aantal_verloren':'Losses','verschil':'GoalDifference','id_ploeg':'StamNumber'})
dim_standings_df['StandingsDayKey'] = dim_standings_df.index
dim_standings_df = pd.merge(dim_standings_df, temp_df, left_on=['StandingsYear', 'StandingsPlayday'], right_on=['seizoen', 'speeldag'])
dim_standings_df = pd.merge(dim_standings_df,dim_date_df[['DateKey','FullDate']],left_on='datum',right_on='FullDate',)
dim_standings_df = dim_standings_df.drop(columns=['datum','FullDate'])
dim_standings_df
```

```python
dim_goal_df = pd.DataFrame()
dim_goal_df['GoalTimeRelative'] = doelpunten_df['goal_time']
dim_goal_df = dim_goal_df.drop_duplicates()
dim_goal_df = dim_goal_df.reset_index(drop=True)
dim_goal_df['GoalKey'] = dim_goal_df.index


dim_goal_df
```

```python
fact_table_match = pd.DataFrame()
fact_table_match = df_total
fact_table_match['datum'] = pd.to_datetime(fact_table_match['datum'])
fact_table_match = fact_table_match.merge(dim_date_df,left_on='datum',right_on='FullDate')

dim_time_df['Time'] = dim_time_df['Hour'].astype(str).str.zfill(2) + ':' + dim_time_df['Minute'].astype(str).str.zfill(2)
fact_table_match = fact_table_match.merge(dim_time_df,left_on='tijd',right_on='Time')
dim_time_df.drop(columns='Time')

fact_table_match = fact_table_match.merge(dim_team_df,left_on='thuisploeg_stamnummer',right_on='StamNumber')
fact_table_match.rename(columns={'TeamKey':'HomeTeamKey'})
fact_table_match = fact_table_match.drop(columns=['TeamName','StamNumber'])
fact_table_match = fact_table_match.merge(dim_team_df,left_on='uitploeg_stamnummer',right_on='StamNumber')
fact_table_match.rename(columns={'TeamKey':'AwayTeamKey'})
fact_table_match = fact_table_match.drop(columns=['TeamName','StamNumber'])


fact_table_match = fact_table_match.drop(columns=['datum','thuisploeg','uitploeg','FullDate','Year','Quarter','Month','Day','Hour','Minute','Time'])
fact_table_match
```
